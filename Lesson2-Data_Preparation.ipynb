```python
import heapq

def paragraph_length_filter(x):

    lines = x['text'].split('\n') #takes each row i of the dataset. eg.dataset['text'][i]
    
    """
    Returns False iff a page has too few lines(less than 3 lines per page) 
    or lines are too short (less than ).
     """

    #len(lines) returns the number of lines in the page. For example, 28 lines in one page
    #[len(line) for line in lines] returns a list of the size of each line in the page. So in this case 28 sizes
    # heapq.nlargest(3,[len(line) for line in lines]) --> this takes the size 3 largest line 
    # if the 3 largest line in the paragraph has a minimum of 3, then remove it.
    
    if(
        len(lines)<3 
        or min(heapq.nlargest(3, [len(line) for line in lines]))<3
        ):
        return False
    
    return True

"""
#Visualize
lines = code_dataset['text'][0].split('\n')
print(f'len(lines):{len(lines)}')
print([len(line) for line in lines])
print(heapq.nlargest(3,[len(line) for line in lines]))

"""
```

## Remove duplicated lines

```python
def find_duplicates(paragraphs):
    """
    Use this function to find the number of repetitions 
    in the paragraphs.
    """
    
    unique_x = set() # a set of distinct elements
    duplicate_chars = 0
    duplicate_elements = 0 

    for element in paragraphs:
        #print("element:",element)
        if element in unique_x :
            #print('Duplicates: ',element)
            duplicate_chars += len(element)
            duplicate_elements +=1
        else:
            #print('Unique:',element)
            unique_x.add(element)
    return duplicate_elements,duplicate_chars

lines = dataset['text'][0]  #this is a class of string. An element is one char. 

"""
print(type(lines))
print(lines)
find_duplicates(lines) #elements from lines are chars
print('--------------------------')
"""


import re
text = re.compile(r"\n{2,}").split(lines.strip()) #this convert a class of string into a list. An element is one paragraph. 
#print(type(text))
#print(text)
#x,y = find_duplicates(text)
#print(f'{x} {y}')


for data in dataset:
    text = re.compile(r"\n{2,}").split(data['text'].strip())
    find_duplicates(text)

```

```python
import re

def paragraph_repetition_filter(x):
    """
    Returns False iff a page has too many repetitions.
    """
    text = x['text']
    paragraphs = re.compile(r"\n{2,}").split(text.strip())
    paragraphs_duplicates, char_duplicates =  find_duplicates(paragraphs)

    if paragraphs_duplicates / len(paragraphs) >0.3:
        return False
    if char_duplicates / len(text) > 0.2 :
        return False
    return True


```

```python
dataset = dataset.filter(
    paragraph_repetition_filter,
    load_from_cache_file=False
)
```

## Quality filter-language

```python
# !pip install fasttext
```

```python
import urllib
from fasttext.FastText import _FastText

def english_language_filter(ds):
    # load language detection model
    model = _FastText('./L2_language_model.bin')
    
    def is_english(x):
        # Predict language of the text and probability
        language, score = model.predict(x['text'].replace("\n", ""))

        language = language[0].split("__")[2]
        return score > 0.4 and language == "en" # change code here if building a model in another language

    ds = ds.filter(is_english, load_from_cache_file=False, num_proc=1)
    return ds

dataset = english_language_filter(dataset)
```
